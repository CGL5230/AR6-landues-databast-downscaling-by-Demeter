### add the AR6 Forest|Managed,Forest|Afforestation and Reforestation, Forest|Forestry|Harvested Area, Forest|Share

```python
landclass_mapping = {
    'Forest|Managed': 'ProtectedUnmanagedForest',
    'Forest|Afforestation and Reforestation': 'ProtectedUnmanagedForest',
    'Forest|Forestry|Harvested Area': 'ProtectedUnmanagedForest',
    'Forest|Share': 'ProtectedUnmanagedForest'
}

grouped =AR6_Argentina_Forest.groupby(['GCAM_basin_ID', 'Variable'])
summed = grouped.sum().reset_index()

# 创建新行
new_rows = []
for landclass, group in summed.groupby('Variable'):
    if landclass in landclass_mapping:
        new_landclass = landclass_mapping[landclass]
        new_row = group.groupby('GCAM_basin_ID').sum().reset_index()
        new_row['Variable'] = new_landclass
        new_rows.append(new_row)


for new_row in new_rows:
    AR6_Argentina_Forest_new  = pd.concat([AR6_Argentina_Forest, new_row], ignore_index=True)

```

```python
AR6_Argentina_Forest_new
```

### calculate the weight for AR6 basin

```python
year_columns = [col for col in Demeter_Argentina_Forest.columns if col not in ['region', 'landclass', 'metric_id'] and 1975 <= int(col) <= 2015]
columns_of_interest = ['region', 'landclass', 'metric_id'] + year_columns

Demeter_Argentina_Forest_sel = Demeter_Argentina_Forest[columns_of_interest]

Demeter_Argentina_Forest_sel['sum'] = Demeter_Argentina_Forest_sel.iloc[:,3:].sum(axis=1)

landclass_totals = Demeter_Argentina_Forest_sel.groupby('landclass')['sum'].sum().reset_index(name='total_sum')

Demeter_Argentina_Forest_sel = Demeter_Argentina_Forest_sel.merge(landclass_totals, on='landclass')

# 计算占比
Demeter_Argentina_Forest_sel['weight'] = Demeter_Argentina_Forest_sel['sum'] / Demeter_Argentina_Forest_sel['total_sum']
```

```python
Demeter_Argentina_Forest_sel
```

### assign the basin value in AR6 data

```python
AR6_Argentina_Forest_allocation = pd.merge(Demeter_Argentina_Forest_sel,AR6_Argentina_Forest_new,left_on=['metric_id','landclass'],right_on=['GCAM_basin_ID','Variable'])
```

```python
AR6_Argentina_Forest_allocation = AR6_Argentina_Forest_allocation.drop(['sum','total_sum','Unnamed: 0','Variable','GCAM_basin_ID','ISO','Country_name','Country','GCAM_Region'],axis=1)
```

```python
AR6_Argentina_Forest_allocation.iloc[:,-17:] = AR6_Argentina_Forest_allocation.iloc[:,-17:] .multiply(AR6_Argentina_Forest_allocation['weight'], axis=0)
```
